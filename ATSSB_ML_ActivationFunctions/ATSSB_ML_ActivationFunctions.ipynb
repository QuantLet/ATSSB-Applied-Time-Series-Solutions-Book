# Activation Functions

**10.1** First plot the sigmoid (with $a=1,2,3$), tanh and ReLU functions using Python, respectively. Then show that $\tanh(s) = 2\,\mathrm{sigmoid}(s/2) - 1$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(s, a=1):
    """Sigmoid function with scaling parameter a."""
    return 1 / (1 + np.exp(-a * s))

def relu(x, a=1):
    """ReLU function with scaling parameter a."""
    return np.maximum(0, a * x)

def tanh_scaled(x, a=1):
    """Hyperbolic tangent function with scaling parameter a."""
    return np.tanh(a * x)
```

```python
s = np.linspace(-10, 10, 200)

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(s, sigmoid(s, a=1), linestyle='-',  label='a=1', linewidth=2, color='#3B75AF')
ax.plot(s, sigmoid(s, a=2), linestyle='--', label='a=2', linewidth=2, color='#EA3728')
ax.plot(s, sigmoid(s, a=3), linestyle=':',  label='a=3', linewidth=2, color='#00A651')

ax.set_xlabel('$s$', fontsize=12)
ax.set_ylabel('Sigmoid function', fontsize=12)
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.2), ncol=3)
ax.set_title('Sigmoid Functions with Different Scaling Parameters')

plt.savefig('sigmoid_function.png', transparent=True, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
x = np.linspace(-10, 10, 200)

fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, relu(x, a=1), linestyle='-',  label='a=1', linewidth=2, color='#3B75AF')
ax.plot(x, relu(x, a=2), linestyle='--', label='a=2', linewidth=2, color='#EA3728')
ax.plot(x, relu(x, a=3), linestyle=':',  label='a=3', linewidth=2, color='#00A651')

ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('ReLU function', fontsize=12)
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.2), ncol=3)
ax.set_title('ReLU Functions with Different Scaling Parameters')
ax.set_ylim(-2, 25)

plt.savefig('relu_function.png', transparent=True, bbox_inches='tight')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))
ax.plot(x, tanh_scaled(x, a=1), linestyle='-',  label='a=1', linewidth=2, color='#3B75AF')
ax.plot(x, tanh_scaled(x, a=2), linestyle='--', label='a=2', linewidth=2, color='#EA3728')
ax.plot(x, tanh_scaled(x, a=3), linestyle=':',  label='a=3', linewidth=2, color='#00A651')

ax.set_xlabel('$x$', fontsize=12)
ax.set_ylabel('Tanh function', fontsize=12)
plt.legend(loc='lower center', bbox_to_anchor=(0.5, -0.2), ncol=3)
ax.set_title('Tanh Functions with Different Scaling Parameters')

plt.savefig('tanh_function.png', transparent=True, bbox_inches='tight')
plt.tight_layout()
plt.show()
```
