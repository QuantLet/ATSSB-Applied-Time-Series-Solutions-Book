```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Defining the time range
t = np.arange(500)

# Defining the functions
f_t = pd.Series(0.2 + 0.1 * t)
p_t = pd.Series(2 * np.sin(2 * np.pi * t / 50 + 0.3 * np.pi))

# Generating noise
np.random.seed(1357)
x_n = pd.Series(np.random.normal(loc=0, scale=4.2, size=500))

# Calculating the required series
fMpMx = f_t * p_t * x_n
```

```python
plt.figure(figsize=(10, 4))
fMpMx.plot()
plt.title(r'$(0.2+0.1t) \cdot 2\sin(2\pi t/50 + 0.3\pi) \cdot N(0, 4.2^2)$')
plt.xlabel('Time')
plt.tight_layout()
plt.savefig("linear_periodic_gaussian_product", transparent=True, dpi=600)
plt.show()
```
