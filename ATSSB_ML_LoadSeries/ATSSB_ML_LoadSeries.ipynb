# Electricity Load Series

Plot the electricity load time series from the `elec-temp.csv` dataset used in Chapter 10 of ATSSB.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('elec-temp.csv', parse_dates=['time'], index_col='time')
df['load'].plot(figsize=(12, 4), title='Electricity Load', color='#3B75AF')
plt.savefig('load_series.png', transparent=True, bbox_inches='tight')
plt.show()
```
