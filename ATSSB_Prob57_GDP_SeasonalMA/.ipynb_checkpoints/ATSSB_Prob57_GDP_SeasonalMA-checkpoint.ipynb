```python
# ==========================================================
# ATSSB  Chapter 5 Production Pipeline
# Clean, reusable, publication-ready
# ==========================================================

from pathlib import Path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

from PythonTsa.ModResidDiag import plot_ResidDiag
from PythonTsa.plot_acf_pacf import acf_pacf_fig
from PythonTsa.datadir import getdtapath

# ==========================================================
# GLOBAL STYLE
# ==========================================================

plt.rcParams.update({
    "figure.figsize": (10,6),
    "figure.facecolor": "none",
    "axes.facecolor": "none",
    "savefig.transparent": True,
    "axes.grid": False,
    "axes.spines.top": False,
    "axes.spines.right": False,
    "legend.frameon": False
})

# ==========================================================
# CORE PIPELINE CLASS
# ==========================================================

class Ch5Pipeline:

    def __init__(self, output_folder="Chapter5_Output"):
        self.root = Path(output_folder)
        self.root.mkdir(exist_ok=True)

    # ------------------------------------------------------
    # Safe figure saving
    # ------------------------------------------------------

    def save_fig(self, name, subfolder=None):
        folder = self.root if subfolder is None else self.root / subfolder
        folder.mkdir(exist_ok=True)

        fig = plt.gcf()
        fig.savefig(folder / f"{name}.png",
                    dpi=300,
                    transparent=True,
                    bbox_inches="tight")
        fig.savefig(folder / f"{name}.pdf",
                    transparent=True,
                    bbox_inches="tight")

    # ------------------------------------------------------
    # Time series plot
    # ------------------------------------------------------

    def plot_series(self, series, name, label="Series", subfolder=None):
        fig, ax = plt.subplots()
        series.plot(ax=ax, label=label)
        ax.legend(loc="upper center",
                  bbox_to_anchor=(0.5, -0.15),
                  ncol=1)
        self.save_fig(name, subfolder)
        plt.show()

    # ------------------------------------------------------
    # ACF / PACF
    # ------------------------------------------------------

    def plot_acf_pacf(self, series, name, lags=36, subfolder=None):
        acf_pacf_fig(series, both=True, lag=lags)
        self.save_fig(name, subfolder)
        plt.show()

    # ------------------------------------------------------
    # Residual diagnostics
    # ------------------------------------------------------

    def plot_diagnostics(self, residuals, name,
                         noestimatedcoef,
                         nolags=28,
                         lag=32,
                         subfolder=None):

        plot_ResidDiag(residuals,
                       noestimatedcoef=noestimatedcoef,
                       nolags=nolags,
                       lag=lag)

        self.save_fig(name, subfolder)
        plt.show()

    # ------------------------------------------------------
    # Fit SARIMA model
    # ------------------------------------------------------

    def fit_sarima(self, series,
                   order,
                   seasonal_order,
                   trend="n",
                   method=None):

        model = sm.tsa.SARIMAX(
            series,
            order=order,
            seasonal_order=seasonal_order,
            trend=trend
        )

        if method:
            fit = model.fit(method=method, disp=False)
        else:
            fit = model.fit(disp=False)

        return fit

    # ------------------------------------------------------
    # Information criteria comparison
    # ------------------------------------------------------

    def compare_models(self, model_dict):

        rows = []

        for name, model in model_dict.items():
            rows.append({
                "Model": name,
                "AIC": model.aic,
                "BIC": model.bic,
                "HQIC": model.hqic
            })

        df = pd.DataFrame(rows).sort_values("AIC")

        print("\n===== MODEL COMPARISON =====")
        print(df.round(3))

        return df


# ==========================================================
# EXAMPLE: PROBLEM 5.7
# ==========================================================

if __name__ == "__main__":

    pipeline = Ch5Pipeline()

    # --------------------------------------------
    # Load data
    # --------------------------------------------

    dtapath = getdtapath()

    x = pd.read_csv(
        dtapath + 'gdpquarterlychina1992.1-2017.4.csv',
        header=0
    )

    dates = pd.date_range(start='1992',
                          periods=len(x),
                          freq='QE')

    x.index = dates
    lx = np.log(pd.Series(x['GDP']))

    pipeline.plot_series(lx,
                         name="log_gdp",
                         label="log(GDP)",
                         subfolder="Prob57")

    # --------------------------------------------
    # Full Model
    # --------------------------------------------

    model_full = pipeline.fit_sarima(
        lx,
        order=(0,1,2),
        seasonal_order=(0,1,3,4),
        trend="n",
        method="powell"
    )

    print("\nFull Model Summary:")
    print(model_full.summary())

    resid_full = model_full.resid[5:]

    pipeline.plot_diagnostics(
        resid_full,
        name="sarima_021_013_diagnostics",
        noestimatedcoef=5,
        subfolder="Prob57"
    )

    # --------------------------------------------
    # Reduced Model
    # --------------------------------------------

    model_reduced = pipeline.fit_sarima(
        lx,
        order=(0,1,2),
        seasonal_order=(0,1,[3],4),
        trend="n"
    )

    print("\nReduced Model Summary:")
    print(model_reduced.summary())

    resid_reduced = model_reduced.resid[5:]

    pipeline.plot_diagnostics(
        resid_reduced,
        name="sarima_021_01_3_diagnostics",
        noestimatedcoef=3,
        subfolder="Prob57"
    )

    # --------------------------------------------
    # Comparison
    # --------------------------------------------

    pipeline.compare_models({
        "Full (0,1,2)(0,1,3)_4": model_full,
        "Reduced (0,1,2)(0,1,[3])_4": model_reduced
    })

    print("\nChapter 5 production pipeline completed.")
```
