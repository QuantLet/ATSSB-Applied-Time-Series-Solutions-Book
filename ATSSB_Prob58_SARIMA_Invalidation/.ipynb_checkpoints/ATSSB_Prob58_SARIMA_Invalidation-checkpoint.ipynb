# ATSSB_Prob58_SARIMA_Invalidation

```python
# ==========================================================
# ATSSB  Problem 5.8 (Saved Charts Version)
# ==========================================================

from pathlib import Path
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt

from PythonTsa.LjungBoxtest import plot_LB_pvalue
from PythonTsa.datadir import getdtapath

# ==========================================================
# Global plot style
# ==========================================================

plt.rcParams.update({
    "figure.figsize": (10,6),
    "figure.facecolor": "none",
    "axes.facecolor": "none",
    "savefig.transparent": True,
    "axes.grid": False,
    "axes.spines.top": False,
    "axes.spines.right": False,
    "legend.frameon": False
})

# ==========================================================
# Output folder
# ==========================================================

FIG_DIR = Path("Prob58_figs")
FIG_DIR.mkdir(exist_ok=True)

def save_current_figure(name):
    fig = plt.gcf()
    fig.savefig(FIG_DIR / f"{name}.png",
                dpi=300,
                transparent=True,
                bbox_inches="tight")
    fig.savefig(FIG_DIR / f"{name}.pdf",
                transparent=True,
                bbox_inches="tight")

# ==========================================================
# Load h02 data
# ==========================================================

dtapath = getdtapath()

h02 = pd.read_csv(
    dtapath + 'h02July1991June2008.csv',
    header=0
)

dates = pd.date_range(start='1991-07',
                      periods=len(h02),
                      freq='ME')

h02.index = dates
h02 = pd.Series(h02['h02'])

# ==========================================================
# Plot series
# ==========================================================

fig, ax = plt.subplots()
h02.plot(ax=ax, label="h02")
ax.legend(loc="upper center",
          bbox_to_anchor=(0.5, -0.15))

save_current_figure("series")
plt.show()

# ==========================================================
# 1 SARIMA(3,0,0)(2,1,0)_12
# ==========================================================

model1 = sm.tsa.SARIMAX(
    h02,
    order=(3,0,0),
    seasonal_order=(2,1,0,12),
    trend='n'
).fit(disp=False)

print("\n===== SARIMA(3,0,0)(2,1,0)_12 =====")
print(model1.summary())

resid1 = model1.resid[12:]

plot_LB_pvalue(resid1,
               noestimatedcoef=5,
               nolags=48)

plt.title("Ljung-Box P-values: (3,0,0)(2,1,0)_12")
save_current_figure("sarima30D20_LB")
plt.show()

# ==========================================================
# 2 SARIMA(3,0,0)(0,1,1)_12
# ==========================================================

model2 = sm.tsa.SARIMAX(
    h02,
    order=(3,0,0),
    seasonal_order=(0,1,1,12),
    trend='n'
).fit(disp=False)

print("\n===== SARIMA(3,0,0)(0,1,1)_12 =====")
print(model2.summary())

resid2 = model2.resid[12:]

plot_LB_pvalue(resid2,
               noestimatedcoef=4,
               nolags=48)

plt.title("Ljung-Box P-values: (3,0,0)(0,1,1)_12")
save_current_figure("sarima30D01_LB")
plt.show()

# ==========================================================
# 3 SARIMA(0,0,5)(0,1,1)_12
# ==========================================================

model3 = sm.tsa.SARIMAX(
    h02,
    order=(0,0,5),
    seasonal_order=(0,1,1,12),
    trend='n'
).fit(disp=False)

print("\n===== SARIMA(0,0,5)(0,1,1)_12 =====")
print(model3.summary())

resid3 = model3.resid[12:]

plot_LB_pvalue(resid3,
               noestimatedcoef=6,
               nolags=48)

plt.title("Ljung-Box P-values: (0,0,5)(0,1,1)_12")
save_current_figure("sarima005D01_LB")
plt.show()

print("\nAll Problem 5.8 charts saved successfully.")
```
