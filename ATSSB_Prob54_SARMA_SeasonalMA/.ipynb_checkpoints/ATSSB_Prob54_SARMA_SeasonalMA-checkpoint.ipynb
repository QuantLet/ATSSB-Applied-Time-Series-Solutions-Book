# ATSSB_Prob54_SARMA_SeasonalMA

```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

from PythonTsa.plot_acf_pacf import acf_pacf_fig
from PythonTsa.ModResidDiag import plot_ResidDiag

plt.rcParams.update({
    "figure.figsize": (10,6),
    "figure.facecolor": "none",
    "axes.facecolor": "none",
    "savefig.transparent": True,
    "axes.grid": False,
    "axes.spines.top": False,
    "axes.spines.right": False,
    "legend.frameon": False
})

# =========================
# PROBLEM 5.4 (CORRECT)
# SARMA(1,0)(0,1)_6
# =========================

FIG_DIR = Path("Prob54_figs")
FIG_DIR.mkdir(exist_ok=True)

# Define SARMA(1,0)(0,1)_6 model
sim_model = sm.tsa.SARIMAX(
    [0],
    order=(1,0,0),
    seasonal_order=(0,0,1,6),
    trend="n"
)

# Parameter order: ['ar.L1', 'ma.S.L6', 'sigma2']
params = np.array([0.6, -0.5, 4.0])

sim = sim_model.simulate(params=params, nsimulations=1000)
simts = pd.Series(sim)

# -------- Series Plot --------
fig, ax = plt.subplots()
ax.plot(simts, label="Simulated series")
ax.legend(loc="upper center", bbox_to_anchor=(0.5,-0.15))
fig.savefig(FIG_DIR/"series.png", dpi=300, transparent=True, bbox_inches="tight")
fig.savefig(FIG_DIR/"series.pdf", transparent=True, bbox_inches="tight")
plt.show()

# -------- ACF / PACF --------
acf_pacf_fig(simts, both=True, lag=30)
plt.savefig(FIG_DIR/"acf_pacf.png", dpi=300, transparent=True, bbox_inches="tight")
plt.savefig(FIG_DIR/"acf_pacf.pdf", transparent=True, bbox_inches="tight")
plt.show()

# -------- Fit Model Back --------
fit = sm.tsa.SARIMAX(
    simts,
    order=(1,0,0),
    seasonal_order=(0,0,1,6),
    trend="n"
).fit(disp=False)

print(fit.summary())

# -------- Diagnostics --------
plot_ResidDiag(fit.resid, noestimatedcoef=2, nolags=30, lag=30)
plt.savefig(FIG_DIR/"diagnostics.png", dpi=300, transparent=True, bbox_inches="tight")
plt.savefig(FIG_DIR/"diagnostics.pdf", transparent=True, bbox_inches="tight")
plt.show()
```
