```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<p><b>2.1.</b> Examine the composition time series that are not examined in Sect. 2.3.</p>

```python
# Defining the time range
t = np.arange(500)

# Defining the functions
f_t = pd.Series(0.2 + 0.1 * t)
p_t = pd.Series(2 * np.sin(2 * np.pi * t / 50 + 0.3 * np.pi))

# Generating noise
np.random.seed(1357)
x_n = pd.Series(np.random.normal(loc=0, scale=4.2, size=500))

# Calculating the required series
fx_n = f_t + x_n
fx = f_t * x_n
px = p_t * x_n
```

```python
plt.figure(figsize=(10, 4))
fx_n.plot()
plt.title(r'$0.2+0.1t + N(0, 4.2^2)$')
plt.xlabel('Time')
plt.tight_layout()
plt.savefig("linear_gaussian_addition", transparent=True, dpi=600)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
fx.plot()
plt.title(r'$(0.2+0.1t) N(0, 4.2^2)$')
plt.xlabel('Time')
plt.tight_layout()
plt.savefig("linear_gaussian_multiplication", transparent=True, dpi=600)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
px.plot()
plt.title(r'$2\sin(2\pi t/50 + 0.3\pi) N(0, 4.2^2)$')
plt.xlabel('Time')
plt.tight_layout()
plt.savefig("periodic_gaussian_multiplication", transparent=True, dpi=600)
plt.show()
```

```python
# Additional calculations
fAp_x = (f_t + p_t) * x_n
fTx_p = (f_t * x_n) + p_t
pAx_f = (p_t * x_n) + f_t
```

```python
plt.figure(figsize=(10, 4))
fAp_x.plot()
plt.title(r'$(0.2+0.1t + 2\sin(2\pi t/50+0.3\pi)) N(0, 4.2^2)$')
plt.xlabel('Time')
plt.tight_layout()
plt.savefig("linear_period_gaussian_multiplication", transparent=True, dpi=600)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
fTx_p.plot()
plt.title(r'$(0.2+0.1t) N(0, 4.2^2) + 2\sin(2\pi t/50+0.3\pi)$')
plt.xlabel('Time')
plt.tight_layout()
plt.savefig("linear_gaussian_period_addition", transparent=True, dpi=600)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
pAx_f.plot()
plt.title(r'$2\sin(2\pi t/50+0.3\pi) N(0, 4.2^2) + (0.2+0.1t)$')
plt.xlabel('Time')
plt.tight_layout()
plt.savefig("period_gaussian_linear_addition", transparent=True, dpi=600)
plt.show()
```

```python
# Defining the time range
t = np.arange(500)

# Defining the functions
f_t = pd.Series(0.2 + 0.1 * t)
p_t = pd.Series(2 * np.sin(2 * np.pi * t / 50 + 0.3 * np.pi))

# Generating noise
np.random.seed(1357)
x_n = pd.Series(np.random.normal(loc=0, scale=4.2, size=500))

# Calculating the required series
fx_n = f_t + x_n
fx = f_t * x_n
px = p_t * x_n

# Additional calculations
fMpMx = f_t * p_t * x_n
fMpAx = (f_t * p_t) + x_n
fMxAp = (f_t * x_n) + p_t
pMxAf = (p_t * x_n) + f_t

# Plotting the graphs
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

fMpMx.plot(ax=axes[0])
axes[0].set_title(r'$(0.2+0.1t) \cdot 2\sin(2\pi t/50 + 0.3\pi) \cdot N(0, 4.2^2)$')

fMpAx.plot(ax=axes[1])
axes[1].set_title(r'$(0.2+0.1t) \cdot 2\sin(2\pi t/50+0.3\pi) + N(0, 4.2^2)$')

plt.xlabel('Time')
plt.tight_layout()
plt.savefig("more_combinations_1", transparent=True, dpi=600)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

fMxAp.plot(ax=axes[0])
axes[0].set_title(r'$(0.2+0.1t) \cdot N(0, 4.2^2) + 2\sin(2\pi t/50+0.3\pi)$')

pMxAf.plot(ax=axes[1])
axes[1].set_title(r'$2\sin(2\pi t/50+0.3\pi) \cdot N(0, 4.2^2) + (0.2+0.1t)$')

plt.xlabel('Time')
plt.tight_layout()
plt.savefig("more_combinations_2", transparent=True, dpi=600)
plt.show()
```
