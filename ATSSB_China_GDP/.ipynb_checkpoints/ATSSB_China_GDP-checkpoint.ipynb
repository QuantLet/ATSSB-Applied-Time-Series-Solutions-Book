```python
import pandas as pd
import matplotlib.pyplot as plt
from PythonTsa.datadir import getdtapath
from statsmodels.graphics.tsaplots import quarter_plot
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.seasonal import seasonal_decompose
from pandas.plotting import lag_plot
from PythonTsa.plot_acf_pacf import acf_pacf_fig
```

```python
# Loading the dataset
dtapath = getdtapath()
x = pd.read_csv(dtapath + 'gdpquarterlychina1992.1-2017.4.csv', header=0)

# Setting the date index
dates = pd.date_range(start='1992', periods=len(x), freq='Q')
x.index = dates
x = pd.Series(x['GDP'], index=dates)

# Calculating moving averages
rollm3 = x.rolling(window=3, center=True).mean()
rollm12 = x.rolling(window=12, center=True).mean()

# Plotting GDP and moving average of order 3
plt.figure(figsize=(10, 6))
plt.plot(x, label='GDP', linestyle='--')
plt.plot(rollm3, label='Moving Average (Order 3)', linestyle='-')
plt.legend(loc="lower center", bbox_to_anchor=(0.5, -0.25), ncol=2, frameon=False)
plt.title('GDP and Moving Average of Order 3')
plt.xlabel('Date')
plt.ylabel('GDP')
plt.savefig("china_gdp_ma_3", transparent=True, dpi=600)
plt.show()
```

```python
# Plotting GDP and moving average of order 12
plt.figure(figsize=(10, 6))
plt.plot(x, label='GDP', linestyle='--')
plt.plot(rollm12, label='Moving Average (Order 12)', linestyle='-')
plt.legend(loc="lower center", bbox_to_anchor=(0.5, -0.25), ncol=2, frameon=False)
plt.title('GDP and Moving Average of Order 12')
plt.xlabel('Date')
plt.ylabel('GDP')
plt.savefig("china_gdp_ma_12", transparent=True, dpi=600)
plt.show()
```

```python
quarter_plot(x)
plt.savefig("gdp_quarterly_seasonal_plot", transparent=True, dpi=600)
plt.show()
```

```python
xhwfit=ExponentialSmoothing(x,trend='mul',seasonal='mul',seasonal_periods=4).fit(method='L-BFGS-B')
# now 'trend' used, rather than 'slope'
# in my opinion, 'slope' is a more appropriate term here.
ax1=plt.subplot(411)
x.plot()
plt.setp(ax1.get_xticklabels(),visible=False)
plt.ylabel('GDP')
ax2=plt.subplot(412,sharex=ax1)
xhwfit.level.plot()
plt.setp(ax2.get_xticklabels(),visible=False)
plt.ylabel('Level')
ax3=plt.subplot(413,sharex=ax1)
xhwfit.trend.plot();plt.setp(ax3.get_xticklabels(),visible=False); plt.ylabel('Trend')
ax4=plt.subplot(414, sharex=ax1)
xhwfit.season.plot()
plt.ylabel('Season')
plt.savefig("exponential_smoothing_dec", transparent=True, dpi=600)
plt.show()
```

```python
xhwfit.resid.plot()
plt.savefig("exponential_smoothing_resid", transparent=True, dpi=600)
plt.show()
```

```python
lag_plot( xhwfit.resid)
plt.savefig("exponential_smoothing_resid_lag", transparent=True, dpi=600)
plt.show()
```

```python
acf_pacf_fig(xhwfit.resid, both=False, lag=20)
plt.savefig("exponential_smoothing_resid_acf", transparent=True, dpi=600)
plt.show()
```

```python
y=xhwfit.forecast(4) #out-of-sample forecasting
y
```

```python
z=xhwfit.predict(start='1992-03-31',end='2018-12-31')
#In-sample prediction and out-of-sample forecasting
z=pd.DataFrame(z,columns=['Predict'])
# now columns=[], not columns={}
zx=z.join(x)
Predict,=plt.plot(zx['Predict'],marker='.',label='Predict')
GDP,=plt.plot(zx['GDP'], linewidth=1.0, label='GDP')
plt.title('Chinese Quarterly GDP and Predict')
plt.legend(handles=[Predict, GDP], loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol=2)
plt.savefig("exponential_smoothing_predict", transparent=True, dpi=600)
plt.show()
```

```python
r,q,p=acf(xhwfit.resid,qstat=True, nlags=20)
p
```

```python
# Seasonal decomposition (Additive Model)
xdeca = seasonal_decompose(x, model='additive')

# Seasonal Adjustment
seasonally_adjusted = x - xdeca.seasonal

# Plotting Seasonally Adjusted Series
seasonally_adjusted.plot(figsize=(10, 6))
plt.title('Seasonally Adjusted GDP')
plt.xlabel('Date')
plt.ylabel('GDP (Seasonally Adjusted)')
plt.grid(True)
plt.show()
```
