```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
```

```python
# Set the parameters
mean = 2.3
std_dev = 1.2
sample_size = 10

# Generate the white noise sample
sample = np.random.normal(mean, std_dev, sample_size)

# Calculate sample mean and standard deviation
sample_mean = np.mean(sample)
sample_std_dev = np.std(sample, ddof=1)  # Using Bessel's correction

# Plot the sample
plt.figure(figsize=(10, 6))
plt.plot(sample, marker='o', linestyle='-', label='White Noise Sample')
plt.axhline(y=sample_mean, color='red', linestyle='--', label=f'Mean = {sample_mean:.2f}')
plt.title('White Noise Sample with Mean')
plt.xlabel('Sample Index')
plt.ylabel('Value')

# Adjust legend position
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2)
plt.grid(False)  # Remove grid
plt.savefig("white_noise_simulation", transparent=True, dpi=600)
plt.show()

# Output the sample mean and standard deviation
(sample_mean, sample_std_dev)
```
